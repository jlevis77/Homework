```python
# 5. Combined and sort the following arrays into one called 'comboArray'
import numpy as np
# list arrays
one = ([10, 11, 12, 13, 14, 15, 16, 17])
two = ([20, 21, 22, 23, 24, 25, 26, 27])
three = ([ 0, 1, 2, 3, 4, 5, 6, 7])

# combine & sort arrays
comboarray = np.concatenate((one, two, three), axis=0)
comboArray = np.sort(comboarray)
print(comboArray)

# 6. Take 'comboArray' and perform the following slicing activities
print("\nLength of comboArray:", len(comboArray))
# print sec1 - the 2nd element
print("\nSec1:", comboArray[1])
# print sec 2 - all elements from 3rd to last
print("\nSec2:", comboArray[2:])
# print sec3 - all elements from the 4th to the 14th elements
print("\nSec 3:", comboArray[3:14])
# print sec4 - last 6 elements
print("\nSec 4:", comboArray[-6:])
# print sec5 - all element from #0 up to and including #15, using the negative number method
print("\nSec 5:", comboArray[:-10])
# print sec6 - from #20 every even element to the end
print("\nSec 6:", comboArray[16::2])
# print sec7 - from the last element moving forward, every 5th element
print("\nSec 7:", comboArray[-1::-5])

# 7. Using simple math operations, broadcast the addition `(+)`, division `(/)`, and multiplication `(*)` operations for these 2 arrays
from numpy import array
# Define arrays
arr1 = array([[[ 0.0,  0.0,  0.10], [10.0, 10.30, 10.01]],
        [[20.0, 20.02, 20.10], [30.0, 31.0, 30.30]],
        [[40.40, 24.0, 40.0], [15.0, 15.5, 25.5]]])
arr2 = array([[17.0, 70.0, 10.7]])

print("\nAddition:\n", arr1 + arr2)
print("\nDivision:\n", arr1 / arr2)
print("\nMultiplication:\n", arr1 * arr2)

# 8. Use the NumPy Arithmetic functions to do the subtraction of the above two arrays in point #7.
print("\nSubtraction:\n", np.subtract(arr1, arr2))



```
